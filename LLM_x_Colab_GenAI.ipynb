```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
import torch
```

```python
# load model mistralai/Mistral-7B-Instruct-v0.2
# tiiuae/falcon-7b-instruct (original)
model = "mistralai/Mistral-7B-Instruct-v0.2"
tokenizer = AutoTokenizer.from_pretrained(model)

falcon_pipeline = transformers.pipeline("text-generation",
                                        model=model,
                                        tokenizer=tokenizer,
                                        torch_dtype=torch.bfloat16,
                                        trust_remote_code=True,
                                        device_map="auto"
                                        )
```

```python
# define completion function
def get_completion_falcon(input):
  # prompt = f"You are an expert Data Analyst. Your job is to explain complex Data concepts in simple words. \n{input}"
  # print(prompt)
  system = f"""
  You are an expert in Data Engineering and Data Analysis with a vast knowledge of databases, Python and SQL.
  You are good at explaining Programming concepts in simple words.
  Help as much as you can.
  """
  prompt = f"#### System: {system}\n#### User: \n{input}\n\n#### Response from model:"
  #print(prompt)
  falcon_response = falcon_pipeline(prompt,
                                    max_length=500,
                                    do_sample=True,
                                    top_k=10,
                                    num_return_sequences=1,
                                    eos_token_id=tokenizer.eos_token_id,
                                    )

  return falcon_response
```

```python
# let's prompt
prompt = "Please write a Code in Python to import sample data from a MySQL Database. Use Streamlit to implement the code in a user interface."
response = get_completion_falcon(prompt)
print(response[0]['generated_text'])
```
